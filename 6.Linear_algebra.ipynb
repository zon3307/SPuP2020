### 벡터 합과 차

* 원소별로 계산 : 다음을 구현하고 싶음
  * [1, 2] 더하기 [3, 4] = [4, 6]
  * [5, 3] 빼기 [1, 7] = [4, -4]

````python
def vector_add(v, w):
    """adds two vectors componentwise"""
    return [v_i + w_i for v_i, w_i in zip(v,w)]
````

````python
vector_add([1,2], [3,4])
````

````python
vector_add([1,2,3], [3,4,5])
````

````python
v = [1, 2, 3]
w = [2, 3, 4]
z = vector_add(v, w)
y = vector_add(z, v)
print(y)
````

````python
def vector_subtract(v, w):
    """subtracts two vectors componentwise"""
    return [v_i - w_i for v_i, w_i in zip(v,w)]
````

````python
vector_subtract([0, 0, 1], [1, 2, 3])
````

### 여러 개의 벡터들의 합

* 벡터들의 리스트가 있을 때, 리스트 내의 벡터들을 원소 별로 합하기

````python
def vector_sum(vectors):
    result = vectors[0]
    for vector in vectors[1:]:
        result = vector_add(result, vector)
    return result
````

````python
vectors = [[1, 2, 3],
           [2, 3, 0],
           [0, 1, -2]]

vector_sum(vectors)
````

````python
from functools import reduce
# 앞의 vector_sum과 같은 기능을 가진다.
def vector_sum(vectors):
    return reduce(vector_add, vectors)
````

````python
vectors = [[1, 2, 3],
           [2, 3, 0],
           [0, 1, -2]]

vector_sum(vectors)
````

````python
# 내부적으로는 다음과 동치
vector_add(vector_add([1, 2, 3], [2, 3, 0]), [0, 1, -2])
````

### 스칼라 곱

* 목표 : 3*[1, 2, 3] = [3, 6, 9]

````python
def scalar_multiply(c, v):
    return [c * vi for vi in v]
````

````python
scalar_multiply(3, [1, 2, 3])
````

* 컴포넌트별 평균 : vector_mean([1, 2], [2, 4], [3, 6]) == [3, 6] 

````python
def vector_mean(vectors):
    n = len(vectors)
    return scalar_multiply(1/n, vector_sum(vectors))
````

````python
vector_mean([[1, 2], [2, 4], [3, 6]])
````

### dot product

* 목표 : 
  * dot( [1, 2, 3], [0, 1, 2]) = sum([1 \* 0, 2 \* 1, 3 \* 2]) = sum([0, 2, 6]) = 8

````python
def dot(v, w):
    """v_1 * w_1 + ... + v_n * w_n"""
    return sum(v_i * w_i for v_i, w_i in zip(v, w))
````

````python
dot([1, 2, 3], [0, 1, 2])
````

### 제곱합

* 벡터 원소들의 제곱의 합

````python
def sum_of_squares(v):
    """v_1 * v_1 + ... + v_n * v_n"""
    return dot(v, v)
````

````python
sum_of_squares([0, 1, 2])
````

* 벡터의 크기(magnitude) : 제곱합의 제곱근

````python
import math
def magnitude(v):
    return math.sqrt(sum_of_squares(v))
````

````python
magnitude([3, 4])
````

````python
magnitude([5, 12])
````

### 벡터 사이의 거리

* 한 벡터에서 다른 벡터를 뺀 후, 크기를 구하는 것과 동일

````python
def distance(v, w):
    return magnitude(vector_subtract(v, w))
````

````python
distance([1, 2], [2, 3])
````

````python
A = np.array([[1, 2], [3, 4]])

A
````

````python
# B는 A를 참조만 한다.
B = A
````

````python
# B의 변화는 A에도 영향을 미침
B[0,0] = 10

B
````

* 만약 이런 현상을 원치 않는다면, ```.copy()```를 이용하여 복사한다.
